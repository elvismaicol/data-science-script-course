## Gráfico de Dispersão com Seaborn

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
path = 'C:/Users/Elvis/Documents/Estudo/UDEMY/Formação Cientista de Dados/12-2021/13.Prática em Python/dados/'
```

```python
co2_file = 'co2.csv'
```

```python
df_co2 = pd.read_csv(path + co2_file)
```

```python
df_co2.head()
```

```python
#Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo Type
srn.scatterplot(df_co2.conc, df_co2.uptake, hue = df_co2.Type)
```

```python
# Selecionando registros específicos da base de dados (Quebec e Mississipi)
q = df_co2.loc[df_co2['Type'] == 'Quebec']
m = df_co2.loc[df_co2['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
#mostrando lado a lado
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1, 2, 2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississippi')
```

```python
ch = df_co2.loc[df_co2['Treatment'] == 'chilled']
nc = df_co2.loc[df_co2['Treatment'] == 'nonchilled']
```

```python
# Gerando gráficos juntos (lado a lado) somente com 'chilled' e 'nonchilled', comparando variáveis caterigoricas e variáveis numeriacas
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(ch.conc, ch.uptake).set_title('Chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(nc.conc, nc.uptake).set_title('Non Chilled')
```
