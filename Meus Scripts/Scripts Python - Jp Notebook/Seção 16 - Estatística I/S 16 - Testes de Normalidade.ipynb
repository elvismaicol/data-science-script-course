## Estatística I - Testes de Distribuição Normal

```python
from scipy import stats
# norm -> para gerar dados normalmente distribuidos
# skewnorm -> para gerar dados com enviesamento(não normais)
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Usando a função "rvs" para criar uma váriavel com dados em um distribuição normmal (de 1000 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
# histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
#Gerando gráfico para verificar se a distribuição é normal
# os pontos de dispersão devem estar junto ao eixo
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
# Executando o teste de Shapiro Wilk (teste de hipótese)
# ficar atento ao valor de "p", se o valor de "p" estiver acima do valor de Alfa(0.05), 
# não há como rejetar a hipótese nula
stats.shapiro(dados)
```

#### Com pvalue=0.43 é maior que 0.05, portanto não pode ser rejeitada a hipótese nula

### Exemplo com Dados não normais

```python
# Gereando dados com enviesament
dados2 = skewnorm.rvs(4, size = 1000)
```

```python
#hitograma
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
#Gerando gráfico para verificar se a distribuição é normal
# os pontos de dispersão não estão junto ao eixo nas bordas(há um alto afastamento)
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
# Executando o teste de Shapiro Wilk (teste de hipótese)
# ficar atento ao valor de "p", se o valor de "p" estiver acima do valor de Alfa(0.05), 
# não há como rejetar a hipótese nula
stats.shapiro(dados2)
```

#### Com pvalue abaixo de 0.05 é rejeitada a hipótese nula
